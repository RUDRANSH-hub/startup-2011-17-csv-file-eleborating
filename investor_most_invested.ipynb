```python
date=list(df.Date)
year=[]
for i in date:
    year.append(i[6:10])
    

#'016',
year
new_yr=[]
for yr in year:
    if len(yr)==3 :
        new_yr.append("2"+yr)
        
    elif yr[0]=="/":
        y=yr.replace("/","")
        new_yr.append(y +"5")  # we can use year[year.index(yr)+1][0] 
    else:
        new_yr.append(yr)
        
print(new_yr)
print(set(new_yr))
```
